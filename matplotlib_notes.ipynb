```python
#example of subplots:
# plot different mathematical functions vs x in a single figure
x = [x * 0.25 for x in range(-25,26)]
pi = math.pi

y1 = [math.sin(x) for x in x]
y2 = [math.cos(x) for x in x]
y3 = [math.sin(x) * -1 for x in x]
y4 = [math.cos(x) * -1 for x in x]

plt.figure(figsize = (8,8))

plt.suptitle('Some Math subplots')

plt.subplot(221)
plt.plot(x,y1)
plt.title('$sin(x)$')

plt.subplot(222)
plt.plot(x,y2)
plt.title('$cos(x)$')

plt.subplot(223)
plt.plot(x,y3,color = 'b', marker = 'o')
plt.title('$-sin(x)$')

plt.subplot(224)
plt.plot(x,y4, color = 'm', marker = '^')
plt.title('$-cos(x)$')
plt.fill_between (x, y4, color = 'm', alpha = 0.7)
#alpha controls transparency. 1 = opaque, 0 = 100% transparency
```

## examples of Histograms

```python
# Explore age of respondents from stackoverflow survey

data = pd.read_csv('https://raw.githubusercontent.com/ravinder79/data/main/age_data.csv', index_col=0)
```

```python
data
```

```python
%matplotlib notebook # allows interaction of the histogram

#range (0,100 and then step of 5 (or spread))
plt.hist(data, bins = range (0, 100, 5))

;
```

```python
#static figure 
%matplotlib inline

median = 29

plt.hist(data,bins = range(0,100,5))
plt.axvline(median, color = 'r', ls = '--',lw = 1, label = 'Median Age');
```

```python

x1 = [random.gauss(0, 0.8) for _ in range(1000)]
x2 = [random.gauss(-2, 1) for _ in range(1000)]
x3 = [random.gauss(3, 1.5) for _ in range(1000)]
x4 = [random.gauss(7, 0.6) for _ in range(1000)]

# make subplots using for loop
plt.figure(figsize = (8,6))
x = [x1, x2, x3, x4] # make a list of lists to iterate through

for i in range(len(x)): 
    plt.subplot(2,2,i+1)# (2 rows, 2 columns, then i + 1 to count out)
    plt.hist(x[i], ec = 'k', color = 'w') #ec is edge color
```

```python
# For bar charts (categorical vs. numeric data)

plt.bar(['cat1', 'cat2','cat3'], [20,30,40])
plt.xlabel('categories')
plt.ylabel('Mean score')
```

```python
plt.barh(['cat1', 'cat2','cat3'], [20,30,40]) # add h in plt.bar(h) for horizontal bars
plt.xlabel('categories')
plt.ylabel('Mean score')
```
