### 1) Use matplotlib to plot the following equation:

```python
#y = x^2 - x + 2
```

#### write the code to generate the x and y points

```python
x = range(-50,50)
y = [(n**2)-(n+2) for n in x]

plt.figure(figsize = (8,6))
plt.plot(x,y)

plt.axvline(x = 0, color ="green", linestyle ="--")

plt.text(0,0,'(0,0)')
```

### 2) Creat and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
#y = sqrt/x
```

```python
Nx = 100
x = np.linspace(0,50,Nx)
y = np.sqrt(x)

plt.figure(figsize = (8,6))
plt.plot(x,y)

plt.axvline(x = 0, color ="green", linestyle ="--")

plt.text(0,0,'(0,0)')
```

```python
#y =x3
```

```python
x = (-50,50)
y = [n * n * n for n in x]

plt.figure(figsize = (8,6))
plt.plot(x,y)

plt.axvline(x = 0, color ="green", linestyle ="--")

plt.text(0,0,'(0,0)')
```

```python
#y = 2 ^ x
```

```python

x = np.arange(10)
y = np.array(2**x)

plt.figure(figsize = (8,6))
plt.plot(x,y)

plt.axvline(x = 0, color ="red", linestyle ="--")

plt.text(0,0,'(0,0)')
```

```python
# y = 1/(x+1)
```

```python
x = np.arange(10)
y = [1/(n+1) for n in x]

plt.figure(figsize = (8,6))
plt.plot(x,y)

plt.axvline(x = 0, color ="red", linestyle ="--")

plt.text(0,0,'(0,0)')
```

### 3) Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = [x * 0.25 for x in range(0,50)]
pi = math.pi

y1 = [(n)**(1./2.)for n in x]
y2 = [n * n * n for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.figure(figsize = (8,8))

plt.suptitle('Finding Squares, Cube and Squareroots')


plt.subplot(221)
plt.plot(x,y1, label="DS learning curve")
plt.legend()
plt.title('$sqrt(x)$')

plt.subplot(222)
plt.plot(x,y2, label="Total Income")
plt.legend()
plt.title('cube(x)$')

plt.subplot(223)
plt.plot(x,y3, label='Gross Sales')
plt.legend()
plt.title('$2(x)$')

plt.subplot(224)
plt.plot(x,y4,label="My brain after JOINS")
plt.legend()
plt.title('$1/(x+1)$')
```
