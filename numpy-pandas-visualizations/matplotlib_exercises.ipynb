### 1) Use matplotlib to plot the following equation:

#### write the code to generate the x and y points

```python
x = range(-50,50)
#let's use for loop. Loop through each element and perform some math
y = [(n**2)-(n+2) for n in x]

plt.figure(figsize = (8,6))
plt.plot(x,y)

plt.axvline(x = 0, color ="green", linestyle ="--")

plt.text(0,0,'(0,0)')

#or can do plt.annotate and in one line add style, line, arrowpoint, etc
```

### 3) Combine the figures you created in the last step into one large figure with 4 subplots.

```python
#generate data
x = [x * 0.25 for x in range(0,50)]
pi = math.pi
x1= range(-50,50)

#y axis'
y1 = [(n)**(1./2.)for n in x]
y2 = [n * n * n for n in x1]
y3 = [2**n for n in x] 
y4 = [1/(n+1) for n in x]


#plot data
plt.figure(figsize = (8,8))
plt.suptitle('Finding Squares, Cube and Squareroots')

#subplots with customizations
plt.subplot(221)
plt.plot(x,y1, label="DS learning curve")
plt.legend()
plt.title('$sqrt(x)$')

plt.subplot(222)
plt.plot(x1,y2, label="Total Income")
plt.legend()
plt.title('cube(x)$')

plt.subplot(223)
plt.plot(x,y3, label='Gross Sales')
plt.legend()
plt.title('$2(x)$')

plt.subplot(224)
plt.plot(x,y4,label="My brain after JOINS")
plt.legend()
plt.title('$1/(x+1)$')
```
