```python
from pydataset import data
import pandas as pd
import numpy as np
from env import get_db_url

url= get_db_url('employees')
```

```python
departments=pd.read_sql('SELECT * FROM departments', url)
dept_emp=pd.read_sql('SELECT * FROM dept_emp ', url)
```

```python
departments
```

```python
titles
```

```python
dept_emp
```

```python
dept_join = dept_emp.merge(departments,
                    left_on='dept_no',
                    right_on='dept_no')
dept_join
```

```python
dept_titles = dept_emp.merge(titles,
                          left_on='emp_no',
                          right_on='emp_no')
dept_titles
```

```python
dept_sections = [dept_titles,dept_join]
result = pd.concat(dept_sections)
result
```

```python
pd.crosstab(result.dept_name, result.title) 
```

```python
result.groupby(['title', 'dept_name']).emp_no.count()
```
