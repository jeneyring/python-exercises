```python
#curves to graph:
## y= 0n + 1 and label the curve "O(1)"
## y =log(n) and label the curve "O(log n)"
## y = n and lable the curve "O(n)"
## y = n * log(n) and label it "O(n log n)"
## y = n^2 and label it "O(n^2)"
## y = 2^n and label it "O(2^n)"
```

```python
x= range(10,50)

plt.figure(figsize = (10,6))

y1 = [(0*n)+1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [(n)**n for n in x]

#customize line plots
plt.plot(x,y1, label ='O(1)')
plt.plot(x,y2, label ='O(log n)')
plt.plot(x,y3, label ='O(n)')
plt.plot(x,y4, label ='O(n log n)')
plt.plot(x,y1, label ='O(n^2)')
plt.plot(x,y1, label ='O(2^n)')
plt.plot(x,y1, label ='O(n!)')
plt.plot(x,y1, label ='O(n^n)')

#add legends
plt.legend(loc='upper right')

#create a title to the lineplot
plt.title('Big O Notation')

#add labels
plt.xlabel('Elements')
plt.ylabel('Operations')

#add grid lines
plt.grid('True', ls = '--')


#add annotations: 
plt.annotate('This Salary', xy = (26, 42000))
```
